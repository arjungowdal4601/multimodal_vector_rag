```python
import os
from pprint import pprint

import chromadb
from dotenv import load_dotenv

load_dotenv()

CHROMA_PATH = os.getenv("CHROMA_PATH", "vector_db/chroma_db")
CHROMA_COLLECTION_NAME = os.getenv("CHROMA_COLLECTION_NAME", "embedding")

print("CHROMA_PATH:", CHROMA_PATH)
print("CHROMA_COLLECTION_NAME:", CHROMA_COLLECTION_NAME)

client = chromadb.PersistentClient(path=CHROMA_PATH)

print("\nAvailable collections:")
collections = client.list_collections()

if not collections:
    print("No collections found")
else:
    for c in collections:
        try:
            print("-", c.name)
        except:
            print("-", c)

collection = client.get_collection(name=CHROMA_COLLECTION_NAME)

print("\nTotal chunks in collection:", collection.count())
```

```python
import chromadb
from pathlib import Path

CHROMA_PATH = "vector_db/chroma_db"

client = chromadb.PersistentClient(path=CHROMA_PATH)

print("Resolved path:", Path(CHROMA_PATH).resolve())
print()

collections = client.list_collections()

for c in collections:
    name = c.name if hasattr(c, "name") else str(c)
    col = client.get_collection(name=name)
    print(f"Collection: {name}")
    print("Count:", col.count())
    print("-" * 60)
```

```python
peek_n = min(5, collection.count())

data = collection.get(
    limit=peek_n,
    include=["documents", "metadatas"]
)

ids = data["ids"]
docs = data["documents"]
metas = data["metadatas"]

for i in range(peek_n):
    print("=" * 100)
    print(f"Chunk {i+1}")
    print("ID:", ids[i])
    print("\nMetadata:")
    pprint(metas[i])
    print("\nText preview:")
    print((docs[i] or "")[:700])
    print()
```
